<h1 style="color:cadetblue; font-size:2em;">Plotting with pandas</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
aapl = pd.read_csv('datasets/AAPL.csv', index_col= 'Date', parse_dates=True)
aapl.head(6)
```

```python
type(aapl.index)
```

```python
close_arr = aapl['Close'].values
type(close_arr)
```

```python
plt.plot(close_arr)
plt.show()
```

```python
close_series = aapl['Close']
type(close_series)
```

<h4 style="text-align:center;">Matplotlip plot</h4>

<h4 style="text-align:center;">Pandas plot</h4>

```python
plt.plot(close_series)
plt.show()
```

```python
close_series.plot()
plt.show()
```

```python
# plotting all Series at once in matplotlib
plt.plot(aapl)
plt.show()
```

```python
# plotting all Series at once in pandas
aapl.plot()
plt.show()
```

<h1 style="color:cadetblue; font-size:2em;">Fixing Scales</h1>

```python
aapl.plot()
plt.yscale('log')  # logaritmic scales on vertical axis
plt.show()
```

<h1 style="color:cadetblue; font-size:2em;">Customizing plots</h1>

```python
aapl['Open'].plot(color='b', style='.-', legend=True)
```

```python
aapl['Close'].plot(color='r', style='.', legend=True)
```

```python
aapl['Open'].plot(color='b', style='.-', legend=True)
aapl['Close'].plot(color='r', style='.', legend=True)
plt.axis(('2015', '2018', 50, 200))
plt.show()
```

<h1 style="color:cadetblue; font-size:2em;">Saving plots</h1>

```python
aapl.loc['2014':'2017', ['Open', 'Close', 'High', 'Low']].plot()
plt.savefig('plots/5.aapl.png')
plt.savefig('plots/5.aapl.jpg')
plt.savefig('plots/5.aapl.pdf')
plt.show()
```
