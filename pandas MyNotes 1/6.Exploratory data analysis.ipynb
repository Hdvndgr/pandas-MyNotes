<h1 style="color:cadetblue; font-size:2em;">Seperating populations</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv('datasets/iris.csv', header=None, names=col_names)
iris.head()
```

```python
iris['species'].describe()
```

```python
iris['species'].unique()
```

```python
indices = iris.species == 'Iris-setosa'
setosa = iris.loc[indices, :]
indices = iris.species == 'Iris-versicolor'
versicolor = iris.loc[indices, :]
indices = iris.species == 'Iris-virginica'
virginica = iris.loc[indices, :]
```

```python
del setosa['species'], versicolor['species'], virginica['species'] 
```

<p>
    <strong>Note:</strong><br>
    If you use setosa.species instead of setosa['species'] you will got an AttributeError.
</p>

```python
setosa.head(2)
```

```python
versicolor.head(2)
```

```python
virginica.head(2)
```

```python
iris.plot(kind='hist', bins=30, range=(0,8), alpha=0.3)
plt.title('Entire iris data set')
plt.xlabel('[cm]')
plt.show()
```

```python
setosa.plot(kind='hist', bins=30, range=(0,8), alpha=0.3)
plt.title('Setosa data set')
plt.xlabel('[cm]')
versicolor.plot(kind='hist', bins=30, range=(0,8), alpha=0.3)
plt.title('Versicolor data set')
plt.xlabel('[cm]')
virginica.plot(kind='hist', bins=30, range=(0,8), alpha=0.3)
plt.title('Virginica data set')
plt.xlabel('[cm]')
plt.show()
```

```python
describe_all = iris.describe()
print(describe_all)
```

```python
describe_setosa = setosa.describe()
describe_versicolor = versicolor.describe()
describe_virginica = virginica.describe()
error_setosa = 100 * np.abs(describe_setosa - describe_all)
error_setosa = error_setosa/describe_setosa
error_versicolor = 100 * np.abs(describe_versicolor - describe_all)
error_versicolor = error_versicolor/describe_versicolor
error_virginica = 100 * np.abs(describe_virginica - describe_all)
error_virginica = error_virginica/describe_virginica
```
