<h1 style="color:cadetblue; font-size:2em;">DataFrame from .csv file without header column</h1>

```python
df2 = pd.read_csv('datasets/file without header.csv', header=None, index_col=0,  na_values={3: -1})
df2
```

Note: By assigning na_values, we decide which values will be showed as Nan value.
names represents column names.
with index_col we decide which column will be index. if we don't assign anything index will be a range of scalar value. 

```python
columns = ['col-1', 'col-2', 'col-3', 'col-4', 'col-5']
df3 = pd.read_csv('datasets/file without header.csv', header=None, names=columns)
df3
```

```python
indexes = ['row_1', 'row_2', 'row_3', 'row_4', 'row_5', 'row_6', 'row_7']
df3.index = indexes
df3
```

Note: df.index changes all indexes, that means if you wanna change only one index name you must change all to do it.
